Faça um boxplot para a variável Schooling. Considere os países:  'Ireland', 'Finland',
'Netherlands', 'Zambia', 'Angola'. Em termos da mediana, qual deles apresenta maior
nível de escolaridade?

```python
df_schooling = [df[df["Country"] == country]["Schooling"] for country in countries]

plt.boxplot(df_schooling, tick_labels = countries)
plt.ylabel("Schooling")
plt.title("Schooling by country")
plt.show()

medians = [float(np.median(each)) for each in df_schooling]
print("\nMedianas:")
for i, country in enumerate(countries):
  print(f' - {country}: {medians[i]}')

print(f'\nAssim, em termos da mediana, o páis que apresenta o maior nível de mediana é {countries[0]} com mediana do nivel de escolaridade de {medians[0]}')
```

Importe os dados do link:  
https://www.worldometers.info/world-population/population-by-country
Construa a matriz de correlação.

Quais as variáveis menos correlacionadas?

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.worldometers.info/world-population/population-by-country/"
# Send a GET request to the URL
response = requests.get(url)
# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
# Find the table containing the population data (find the first table)
table = soup.find('table')
# Convert the table to a Pandas DataFrame
df = pd.read_html(str(table))[0]
# Print the Data

df.head()

df_relevant = df.iloc[:, 2:]
df_relevant = df_relevant.replace("−", "-", regex=True)
df_relevant = df_relevant.replace(",", "", regex=True)

df_relevant['Yearly Change'] = df_relevant['Yearly Change'].str.rstrip('%').astype(float) / 100
df_relevant['Urban Pop %'] = df_relevant['Urban Pop %'].str.rstrip('%').astype(float) / 100
df_relevant['World Share'] = df_relevant['World Share'].str.rstrip('%').astype(float) / 100
df_relevant = df_relevant.astype(float)

corr = df_relevant.corr()
display(corr)



plt.imshow(corr, cmap = "Blues")
plt.xticks(range(len(corr)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr)), corr.columns);
plt.show()

min_value = np.min(corr)

min_position = corr.stack().idxmin()
row, column = min_position

print(f"\nAs variáveis com menos correlação entre si são {row} e {column}")
```
